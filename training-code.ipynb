<h3>Initial Cleaning Of Dataframe.</h3>

```python
import pandas as pd
df=pd.read_csv("gender.csv")
print(df.columns)
```

```python
# Removing Unwanted Column
df.drop(df.columns[-1],axis=1,inplace=True)
print(df.columns)
```

```python
#Looking for anomalies

str_cols=[' Gender' ,' Occupation',' Education Level', ' Marital Status',
       ' Favorite Color']

for i in str_cols:
 print(df[i].unique())
```

```python
#Stripping the values

for i in str_cols:
 df[i] = df[i].apply(lambda x: x.strip())
 print(df[i].unique())
```

```python
# Extracting the labels

df[" Gender"].replace({"female":0,"male":1},inplace=True)
labels=df[" Gender"]
df.drop(" Gender",axis=1,inplace=True)
```

<h3>Applying The Transform (We have created it earlier)</h3>

```python
from transform_module import transform

features=transform(dataframe=df,scaler_filepath="preprocessing_files/robust_scaler.pkl",
                   encoder_filepath="preprocessing_files/onehot_encoder.pkl",pca_filepath="preprocessing_files/pca.pkl")
```

<h3>Training the base model</h3>

```python
from sklearn.model_selection import train_test_split

X_Train,X_Test,Y_Train,Y_Test=train_test_split(features,labels,test_size=0.2)
```

```python
from tensorflow import keras
```

```python
model=keras.Sequential([
 keras.layers.Input(shape=(20,)),
 keras.layers.Dense(8,activation="relu"),
 keras.layers.Dropout(0.2),
 keras.layers.Dense(1,activation="sigmoid")

])
```

```python
model.compile(optimizer="adam",loss="binary_crossentropy",metrics=["accuracy"])
```

```python
model.fit(X_Train,Y_Train,epochs=500)
```

```python
model.evaluate(X_Test,Y_Test)
```

```python
model.save("model_files/base_model.h5")
model.save_weights("model_files/base_weights.h5")
```

<h3>Applying Quantization-Aware-Training and Post-Training Quantization</h3>

```python
import tensorflow_model_optimization as tfmot
quantize_model= tfmot.quantization.keras.quantize_model

model_to_quantize=model
model_to_quantize.load_weights("model_files/base_weights.h5")

q_model=quantize_model(model_to_quantize)

q_model.compile(optimizer="adam",loss="binary_crossentropy",metrics=["accuracy"])

q_model.summary()
```

```python
q_model.fit(X_Train,Y_Train,epochs=500)
```

```python
q_model.evaluate(X_Test,Y_Test)
```

```python
q_model.save("model_files/quantization_aware_model.h5")
```

```python
import tensorflow as tf

def convert_tflite(model,save_location,quantize=False):
    converter= tf.lite.TFLiteConverter.from_keras_model(model)

    if quantize:
        converter.optimizations = [tf.lite.Optimize.DEFAULT]

    tflite_model= converter.convert()

    with open(save_location,"wb") as f:
        f.write(tflite_model)
        
```

```python
convert_tflite(q_model,"model_files/quantization_aware_quantized_model.tflite",quantize=True)
```
